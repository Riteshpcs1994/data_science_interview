```python
# Merge Strings Alternately

def merge_str(s1,s2):
    s3 = []
    for i, z in zip(s1,s2):
        s3.append(i)
        s3.append(z)

    s3.extend(s1[len(s1):])
    s3.extend(s2[len(s2):])

    return "".join(s3)
```

```python
merge_str("rit", "kinshu")
```

```python
# Greatest Common Divisor of Strings

def gcd(a, b):
    while b:
        a, b = b, a%b
    return a

def cds(str1, str2):

    if str1 + str2 != str2 + str1:
        return ""
    
    len_gcd = gcd(len(str1), len(str2))

    return str1[:len_gcd]
    
```

```python
cds("abab", "ab")
```

```python
def rv(str1):
    str1 = list(str1)
    v = "aeiouAEIOU"
    r = len(str1) -1
    l = 0

    while l < r :
        while l < r and str1[r] not in v :
            r -=1
        while l <r and str1[l] not in v:
            l +=1

        str1[l], str1[r] = str1[r], str1[l]

        r -=1
        l +=1
    

    return "".join(str1)
```

```python
rv("IceCreAm")
```

```python
 #Product of Array Except Self

def pes(array1):
    n = len(array1)
    result = [1] * n
    prefix =1

    for i in range(n):
        result[i] =prefix
        prefix *= array1[i]

    
    sufix =1

    for i in range(n-1, -1, -1):
        result[i] *= sufix
        sufix *= array1[i]

    return result
```

```python
nums = [1,2,3,4]

pes(nums)
```

```python
 #Increasing Triplet Subsequence

def its(nums):
    first = float("inf")
    second = float('inf')

    for i in nums :
        if i<= first:
            first = i
        elif i<= second:
            second = i

        else :
            return True

    return False
```

```python
its([5,4,3,2,1])
```

```python
# Max Number of K-Sum Pairs

def mkp(num, k):
    
    num.sort()

    cunt =0
    l = 0
    r = len(num) -1

    while l < r:
        prs = num[l]  + num[r]

        if prs ==k:
            cunt +=1
            l +=1
            r -=1
        if prs < k :
            l +=1
        
        else :
            r -=1

    return cunt


nums = [1,2,3,4]
k = 5
mkp(nums, k)
```

```python
 #Container With Most Water
def mxa(h):

    a= 0
    l = 0
    r = len(h) -1

    while l < r :
        w = r -l
        mn_h = min(h[l], h[r])
        a = max(a, mn_h * w)

        if h[l] <h[r]:
            l +=1
        
        else:
            r -=1
    return a

height = [1,8,6,2,5,4,8,3,7]
mxa(height)
```
