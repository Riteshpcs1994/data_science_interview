2. Write a program in Python to check whether an integer is Armstrong number or not.
    * It is a number which is equal to the sum of cube of its digits.
    * For eg: 153, 370 etc.


```python
# Method 1
def armstrong_check(original_number):
    print(f"Original Number: {original_number}" )
    number = original_number
    number_sum = 0

    while number !=0:
        digit = (number % 10)
        number_sum +=  digit**3
        number = number//10

    if original_number == number_sum:
        return "Yes Number is Armstrong"
    else:
        return "No Armstring"
    
# method 2 
def armstrong_check_str(orginal_number):
    number = str(orginal_number)
    if orginal_number == sum(int(i)**3 for i in number):
        return f"Yes Number {orginal_number} is Armstrong"
    
    else :
        return f"No Number {orginal_number} is not Armstrong"


print(armstrong_check(153))
print(armstrong_check_str(153))
print(armstrong_check(245))
print(armstrong_check_str(245))
```

5) Write a program in Python to check whether a number is palindrome or not using iterative method.

itself.
    * e,g = 121, 1223221

```python
# Method 1
def itrative_palindrome(number):
    if len(str(number)) ==1:
        return "Yes Palindrome"
    original_number = number
    new_number = 0
    while number != 0:
        new_number = new_number*10 + number % 10
        number = number // 10

    if original_number == new_number:
        return "Yes Palindrome"
    else:
        return "No Palindrome"
    
# Method 2
def str_palitrom(number):
    if int(str(number)[::-1]) == number:
        return "Yes Palindrome"
    else:
        return "No Palindrome"
    
# Method 3 recursive
def reverse(num):
    if num < 10:
        return num
    else:
        return (int(str(num % 10) + str(reverse(num//10))))
    

def reverse_palidrom(num):
    if num == reverse(num):
        return "Yes number is Palidrom"
    
    else :
        return "Number is not Palidrom"
    

print(reverse_palidrom(121))
print(itrative_palindrome(1))
print(str_palitrom(1))
```
