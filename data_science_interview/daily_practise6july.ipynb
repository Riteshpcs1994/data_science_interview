##### 1. Given a list of characters, compress it by replacing consecutive repeating characters with the character followed by its count. If a character appears only once, keep only the character (do not add 1). Return the compressed string

```python
def compress_string(chars):
    chars.append(" ")
    compressed = []
    cnt = 1

    for i in range(1, len(chars)):
        if chars[i] == chars[i - 1]:
            cnt += 1
        else:
            compressed.append(chars[i - 1])
            if cnt > 1:
                compressed.append(str(cnt))
            cnt = 1

    return "".join(compressed)

print(compress_string(["a","a","b","b","c","c","c","a"]))

print(compress_string(["a","a","b","b","c","c","c"]))

```

#### 2. Given two arrays representing the arrival and departure times of trains at a railway station, determine the minimum number of platforms required so that no train has to wait.

```python
def min_platforms(arrival, departure):
    arrival.sort()
    departure.sort()

    i = j= 0
    platforms = 0
    max_platforms = 0

    while i < len(arrival) and j < len(departure):
        if arrival[i] < departure[j]:
            platforms +=1
            max_platforms = max(max_platforms, platforms)
            i +=1

        else :
            platforms -=1
            j +=1


    return max_platforms


arrival = []

arrival = [2.00, 2.10, 3.00, 3.20, 3.50, 5.00]
departure = [2.30, 3.40, 3.20, 4.30, 4.00, 5.20]

print(min_platforms(arrival, departure)) 
```
