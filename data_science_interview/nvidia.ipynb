```python
def mnp(ta,td):
    plt_cnt = 1
    d = 0
    a = 0
    plt = 1
    while a < len(ta) :
        if ta[a] <= td[d] :
            plt +=1
            a +=1
        else :
            d +=1
            plt -=1
    return plt
```
