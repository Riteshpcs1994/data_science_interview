```python
import os
import pandas as pd

dir_name = "path_to_your_directory"  # Set your folder path here

all_files = [f for f in os.listdir(dir_name) if f.endswith(".csv")]

# List to store each dataframe
df_list = []

for file in all_files:
    file_path = os.path.join(dir_name, file)
    df = pd.read_csv(file_path)
    df_list.append(df)

# Concatenate all dataframes
new_file = pd.concat(df_list, ignore_index=True)

# Optional: Save the merged file
new_file.to_csv("merged_file.csv", index=False)
```
