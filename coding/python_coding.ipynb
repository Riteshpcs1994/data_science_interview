```python
def sen_highest(input_string):

    dict_str = {}

    for i in input_string:
        if i in dict_str.keys():
            dict_str[i] +=1
        else :
            dict_str[i] =1

    first_cnt = snd_cnt = 0
    first_chr = snd_chr =  None

    for k, v in dict_str.items():
        if v > first_cnt:
            first_cnt = v
            first_chr = k

        elif v > snd_cnt and v < first_cnt:
            snd_cnt = v
            snd_chr = k

            snd_chr = k
    return snd_chr          


    
```
