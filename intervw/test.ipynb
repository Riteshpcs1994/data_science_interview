```python
# i/p: aaaabbbccaa
# o/p: a4b3c2a2
```

```python
def chr_cnt(input_str):
    input_str = input_str + " "
    inc = input_str[0]
    lst = []
    cnt =1
    for i in range(1,len(input_str)):
        if input_str[i] == inc:
            cnt +=1 
        else :
            lst.append(inc)
            lst.append(str(cnt))
            cnt =1
            inc = input_str[i]

    return "".join(lst)

```

```python
chr_cnt("aaaabbbccaa")
```
